# K-S Test

```python
import numpy as np
import seaborn as sb
from scipy import stats
import matplotlib.pyplot as plt

#generate 1000 normal random samples
x=stats.norm.rvs(size=1000)
sb.set_style("whitegrid")

#plot kernel density plot
#variance in kernel is also called bandwidth
sb.kdeplot(np.array(x), bw=0.5)
plt.show()
```

```python
stats.kstest(x, 'norm')
#result contains the ks-test statictic, D and the p-value
```

```python
y=np.random.uniform(0,1,10000)
sb.kdeplot(np.array(y), bw=0.1)
plt.show()
```

```python
stats.kstest(y,'norm')
#result contains ks-test statistic and p-value
#p-value is very very less so it is not normally distributed
#p-value<alpha - reject the null hypothesis which is assuming that the distribution is gaussian
```
